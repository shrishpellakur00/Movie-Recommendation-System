## Movie Recommendation System based on item collaborative filtering

1) Import the libraries

2) Import the dataset

3) Perform item based collaborative filtering on one of the movie samples

4) Visualize the data

5) Create an item based collaborative filtering model for the entire dataset



```
# Import the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# import the data

df = pd.read_csv('Movie_Id_Titles')
```

```
df
```

```
df2 = pd.read_csv('u.data', sep = '\t', names = ['user_id','item_id','rating','timestamp'])
```

```
df2
```

```
# Dropping the timestamp column

df2.drop(['timestamp'], axis=1, inplace=True)
```

```
# Merging the two dataframes together

movies = pd.merge(df2, df, on='item_id')
```

```
movies
```

```
movies.groupby('title').describe()
```

```
movies.groupby('title')['rating'].describe()
```

```
rating_count = movies.groupby('title')['rating'].describe()['count']
```

```
rating_mean = movies.groupby('title')['rating'].describe()['mean']
```

```
rating_mean_count = pd.concat([rating_count, rating_mean], axis = 1)
```

```
rating_mean_count.reset_index()
```

```
rating_mean_count['count'].plot(bins=100, kind='hist', color='r');
```

```
rating_mean_count['mean'].plot(bins=100, kind='hist', color='b');
```

```
rating_mean_count.sort_values('count', ascending = False).head(100)
```

```
userid_matrix = movies.pivot_table(index='user_id', columns='title',values='rating')
```

```
userid_matrix
```

```
titanic = userid_matrix['Titanic (1997)']
```

```
titanic
```

```
titanic_correlations = pd.DataFrame(userid_matrix.corrwith(titanic), columns=['correlations'])
```

```
titanic_correlations = titanic_correlations.join(rating_mean_count['count'])
```

```
titanic_correlations.dropna(inplace=True)
```

```
titanic_correlations
```

```
titanic_correlations.sort_values('correlations',ascending=False)
```

```
titanic_correlations[titanic_correlations['count']>80].sort_values('correlations', ascending=False).head(100)
```

```
# Let's create our own dataframe with our own ratings!
myRatings = pd.read_csv("My_Ratings.csv")
```

```
movie_correlations = userid_matrix.corr(method = 'pearson', min_periods = 80)
# pearson : standard correlation coefficient
# Obtain the correlations between all movies in the dataframe
```

```
myRatings
```

```
similar_movies_list = pd.Series()
for i in range(0, 2):
    similar_movie = movie_correlations[myRatings['Movie Name'][i]].dropna() # Get same movies with same ratings
    similar_movie = similar_movie.map(lambda x: x * myRatings['Ratings'][i]) # Scale the similarity by your given ratings
    similar_movies_list = similar_movies_list.append(similar_movie)
```

```
similar_movies_list.sort_values(inplace = True, ascending = False)
print (similar_movies_list.head(10))
```
